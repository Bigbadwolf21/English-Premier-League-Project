#### Building a Model on the data to predict the full result of games in English Premier League.

```python
# Merging the data together
model_data = data_1.append(data_2)
model_data.head()
```

```python
# The shape of the data
model_data.shape
```

```python
# Columns of the data
model_data.columns
```

```python
# Dropping columns that won't be useful to our model
model_data.drop(['Div', 'Date', 'Time', 'HomeTeam', 'AwayTeam', 'Referee' ], axis = 1, inplace=True)
```

```python
model_data
```

```python
#model_data['FTR'] = model['FTR'].replace()
```

```python
# Let's encode our target variable and 
from sklearn.preprocessing import OneHotEncoder
```

```python
code = OneHotEncoder(handle_unknown='ignore')
coded_data = pd.DataFrame(code.fit_transform(model_data[['HTR']]).toarray())
coded_data
```

```python
# Merge the encoded data with the main data
model_data = model_data.join(coded_data) 
model_data
```

```python
# Drop the HTR Column
model_data.drop(['HTR'], axis=1, inplace=True)
```

```python
# View the model data
model_data
```

```python
# Let's label encode the target variable FTR
# Import label encoder
from sklearn import preprocessing
label_encode = preprocessing.LabelEncoder()
# Encode the FTR Column
model_data['FTR'] = label_encode.fit_transform(model_data['FTR'])

model_data['FTR'].head()
```

```python
model_data
```

```python
np.any(np.isnan(model_data))
```

```python
np.all(np.isfinite(model_data))
```

```python
np.any(np.isnan(model_data)).sum()
```

```python
#np.any(np.isnan(model_data))
model_data = model_data.dropna(how='any')
```

```python

# Setting the variables to target variable and labels
X = model_data.loc[:, ~model_data.columns.isin(['FTR'])]
y = model_data['FTR']
```

```python
# Import the libraries that will be used to build the model
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into traaining set and testing set
# The model will train on the 80% of the data and will test with the remaining 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 , random_state=15)
```

```python
classifier = RF(n_estimators=500, criterion='entropy', random_state=1, max_depth=15, min_samples_split=2, min_samples_leaf=1)
classifier.fit(X_train,y_train)
y_Pred = classifier.predict(X_test)
msquared = mean_squared_error(y_test,y_Pred)
r = r2_score(y_test,y_Pred)
m_abs_error = mean_absolute_error(y_test,y_Pred)
accuracy = accuracy_score(y_test,y_Pred)
print("Random Forest Classifier :")
print("Accuracy = ", accuracy)
print("Mean Squared Error:",msquared)
print("R score:",r)
print("Mean Absolute Error:",m_abs_error)
```
